```python
import json
import datetime

s3 = session.resource('s3')

i = 16
 
while i<=31:
    key = 'agromonitoring/2019/01/' + str(i) + '/metadata/satellite_imagery.json'
    content_object = s3.Object('nestlogic-agro', key)
    file_content = content_object.get()['Body'].read().decode('utf-8')
    json_content = json.loads(file_content)
    print '\n' + '2019/01/' + str(i) + '\n'
    for elem in json_content:
        print datetime.datetime.fromtimestamp(elem.get(u'dt')).isoformat()
    i += 1

```
